```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/train_df_fold.csv")
df
```

```python
# labelの割合を確認
sample = df["label"].value_counts(normalize=True)
# label順に並び替え
sample = sample.sort_index()
sample
```

* label の 1 と 3 の Accuracy が低いのは、サンプル数が少ないことが関係していると考えられる
* 理想：labe1 を 2 倍、label3 を 3 倍に増やす

```python
# foldごとのlabelの割合を確認
sample = df.groupby("fold")["label"].value_counts(normalize=True)
# fold順に並び替え
sample = sample.sort_index()
sample
```

全体の割合と乖離してはいない

```python
# label1の連続数を確認
label1_cont = []
counter = 0
# dfを1つずつ見ていく
for i in range(len(df)):
    # labelが1の場合
    if df.iloc[i]["label"] == 1:
        counter += 1
    else:
        if counter != 0:
            label1_cont.append(counter)
        counter = 0
print(label1_cont)
```

label1 は単発で出現するものが多い

```python
n_label1 = sum(label1_cont)
n_label1
```

```python
# label1_contの1を全てlistから削除
label1_cont = [i for i in label1_cont if i != 1]
print(label1_cont)
```

```python
print(sum(label1_cont))
# 中間にまたがってデータを作成した場合、何個データが増えるか = 15secずらし
increment = sum(label1_cont) - len(label1_cont)
print(f"増分: {increment}")
print(f"増加率: {increment / n_label1}")
```

```python
# 10secずらしだといくつ増えるか
increment = (sum(label1_cont) - len(label1_cont)) * 2
print(f"増分: {increment}")
print(f"増加率: {increment / n_label1}")
```

```python
# label3の連続数を確認
label3_cont = []
counter = 0
# dfを1つずつ見ていく
for i in range(len(df)):
    # labelが1の場合
    if df.iloc[i]["label"] == 3:
        counter += 1
    else:
        if counter != 0:
            label3_cont.append(counter)
        counter = 0
print(label3_cont)
```

```python
n_label3 = sum(label3_cont)
n_label3
# label3_contの1を全てlistから削除
label3_cont = [i for i in label1_cont if i != 1]
print(label3_cont)
```

```python
print(sum(label3_cont))
# 中間にまたがってデータを作成した場合、何個データが増えるか = 15secずらし
increment = sum(label3_cont) - len(label3_cont)
print(f"増分: {increment}")
print(f"増加率: {increment / n_label3}")
```

```python
# 10secずらしだといくつ増えるか
increment = (sum(label3_cont) - len(label3_cont)) * 2
print(f"増分: {increment}")
print(f"増加率: {increment / n_label3}")
```

10sec ずらしてデータを作成することで以下の通りデータを増やすことができる

* label1: 2.4 倍
* label3: 3.3 倍

※ データの連続性を考慮しなかった（異なる被験者データで連続しているかどうか確認していない）ので、実際はもう少し少なくなるかも
