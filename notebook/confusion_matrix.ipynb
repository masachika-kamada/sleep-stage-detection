```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cm = [
    [71409,  1379,  1184,    22,   208],
    [ 5406, 22394,  4503,   415,  1110],
    [ 3659,  2113, 38980,   753,  1513],
    [ 1048,   369,  3066, 17244,    38],
    [ 2478,  1292,  2262,    41, 11922]
]
cm = pd.DataFrame(data=cm, index=["0", "1", "2", "3", "4"], columns=["0", "1", "2", "3", "4"])
sns.heatmap(cm, square=True, cbar=True, annot=True, cmap='Blues')
fig, ax = plt.subplots()
plt.yticks(rotation=0)
plt.xlabel("Pred", fontsize=13, rotation=0)
plt.ylabel("GT", fontsize=13)
ax.set_ylim(len(cm), 0)
plt.show()
```

```python
# ground truth
cm_normalized = cm.div(cm.sum(axis=1), axis=0)

# 
sns.heatmap(cm_normalized, square=True, cbar=True, annot=True, cmap='Blues')
plt.yticks(rotation=0)
plt.xlabel("Pred", fontsize=13, rotation=0)
plt.ylabel("GT", fontsize=13)
ax.set_ylim(len(cm_normalized), 0)
plt.savefig('sklearn_confusion_matrix_normalized.jpg')
plt.show()
```
